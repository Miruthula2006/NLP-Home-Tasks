```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.chunk import tree2conlltags
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from nltk.metrics import precision

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('stopwords')

news = "Microsoft announced a new AI platform in Chennai. The company expects the technology to improve business productivity."

#Parsing Sentence Structure
tokens = word_tokenize(news)
pos_tags = pos_tag(tokens)

print("PARSING:")
print("POS Tags:")
print(pos_tags)

#Named Entity Recognition
print("\nNAMED ENTITIES:")
named_entities = ne_chunk(pos_tags)

for chunk in named_entities:
    if hasattr(chunk, 'label'):
        entity = " ".join(c[0] for c in chunk)
        print(chunk.label(), ":", entity)

#Text Classification
def word_features(text):
    return {word.lower(): True for word in word_tokenize(text)}

train_data = [
    (word_features("team won cricket match"), "Sports"),
    (word_features("government election parliament"), "Politics"),
    (word_features("stock market business profit"), "Business"),
    (word_features("AI technology software innovation"), "Technology")]

classifier = NaiveBayesClassifier.train(train_data)
category = classifier.classify(word_features(news))
print("\nCATEGORY:")
print(category)

#Model Evaluation
test_data = [
    (word_features("AI software platform"), "Technology"),
    (word_features("business market growth"), "Business")]

acc = accuracy(classifier, test_data)
print("\nMODEL ACCURACY:")
print(round(acc * 100, 2), "%")

# Precision Example
reference = {'Technology'}
test = {category}

try:
    prec = precision(reference, test)
    print("Precision:", round(prec, 2))
except:
    print("Precision: Not Calculated")

#Word Frequency & Probability Distribution
words = [w.lower() for w in tokens if w.isalpha()]
freq = FreqDist(words)

print("\nWORD FREQUENCY:")
for word, count in freq.items():
    print(word, ":", count)

total_words = len(words)
print("\nPROBABILITY DISTRIBUTION:")
for word, count in freq.items():
    print(f"P({word}) = {count}/{total_words} = {count/total_words:.2f}")

#Semantic Analysis
print("\nSEMANTIC ANALYSIS:")
important_words = ["AI", "technology", "business"]

for word in important_words:
    print(word, "-> Related to innovation and productivity")

#Bigrams and Trigrams
print("\nBIGRAMS:")
for bg in bigrams(words):
    print(bg)

print("\nTRIGRAMS:")
for tg in trigrams(words):
    print(tg)
```
